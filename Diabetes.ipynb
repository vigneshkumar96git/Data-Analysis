**Predicting the likelihood of diabetes based on age using Logistic Regression**

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
```

```python
df=pd.read_csv('diabetes.csv')
df.head()
```

```python
plt.scatter(df.age,df.diabetes,color='red',marker='+')
```

```python
from sklearn.model_selection import train_test_split
train_X,test_X,train_Y,test_Y=train_test_split(df[['age']],df.diabetes,test_size=0.2,random_state=15)
train_X
```

```python
r=LinearRegression()
r.fit(train_X,train_Y)
r.predict([[35]])
```

```python
plt.scatter(df.age,df.diabetes,color='red',marker='+')
plt.plot(df.age,r.predict(df[['age']]),color='blue')
plt.xlabel("Age")
plt.ylabel("Diabetes")
```

**Logistic Regression**

```python
m=LogisticRegression()
m.fit(train_X,train_Y)
```

```python
prediction_data = pd.DataFrame([[35]], columns=train_X.columns)
m.predict(prediction_data )
```
